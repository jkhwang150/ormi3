```python
# 문제 1. 마지막 문자열 두번더하기
s = input()
answer = s + s[-1] + s[-1]
print(answer)
```

```python
# 문제 2. 대문자를 소문자로 출력
s = input()
answer = s.lower()
print(answer)
```

```python
# 문제 3. 'abcde', 숫자, 대쉬(-), 언더스코어(_) 제거하기
s = input()
rule = str.maketrans('','','abcde1234567890-_')
answer = s.translate(rule)
print(answer)
```
