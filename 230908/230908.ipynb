# 정수 자료형

```python
# 진수
x = 26
print(bin(x)) # 2진수
print(oct(x)) # 8진수
print(hex(x)) # 16진수
```

```python
# 매우 중요한 내용, 새로운 자료형을 만났을 때 2개를 꼭 해보세요.
x = 10
print(type(x)) # 검색 키워드
print(dir(x)) # 대략적인 속성을 알 수 있습니다.
```

```python
# 문제1
x = 100
y = 100
z = x

# 다음 화살표를 그려보세요!
# z ┐
# x -> 100
# y ┘

# 문제1
x = 258
y = 258
z = x

# 다음 화살표를 그려보세요!
# z ┐
# x -> 258
# y -> 258
print(id(x), id(y), id(z))
```

```python
# 문자열 인덱싱
s = 'hello world'
s[0]
s[-1]
s[-11]
```

```python
print(s[0:5])
print(s[:5])
print(s[6:])
print(s[0:5:2])
# 2칸씩 건너 띄는 코드는 거의 작성하지 않고(알고리즘 문제에서는 많이 사용합니다.)
# step 값으로는 -1을 많이 사용합니다.
print(s[::-1])
print(s[5:0:-1]) # ????, 왜 의도한 바대로 안나왔을까요? 5부터 1까지이기 때문에 그렇습니다!
```

```python
# 문자열 메서드
type(s), dir(s)
```

```python
# 대소문자 변환
s.lower() #전체를 소문자로 바꿔주는 method
s.upper() #전체를 대문자로 바꿔주는 method
```

```python
# 특정 인덱스 찾기
s.find('he'), s.index('world') # 위치를 찾아 출력
```

```python
# Count : 특정 문자열의 개수
print(s.count('o'))
# strip : 특정 문자 제거
print(s.strip())
```

```python
# replace: 문자열 변환
s.replace('hello','bye')
```

```python
#split / join : 문자열 합치기 분리하기
print(s.split())
t = s.split()
print('-'.join(s))
```

```python
# zfill : 칸수 맞춰주기
x = "42"
result = x.zfill(5)
print(result)  # 00042
```

```python
# 포멧
년 = '23'
월 = '9'
일 = '8'
오늘날짜 = f'{년}/{월}/{일}'
print(오늘날짜)
오늘날짜 = f'{년}/{월.zfill(2)}/{일.zfill(2)}' #원하는 양식 23/09/08
print(오늘날짜)
```
