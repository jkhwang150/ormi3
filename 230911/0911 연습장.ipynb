```python
# 문제 3. 어떤 email을 사용하는지 구하시오.
email2 = 'abc@gmail.com' # gmail
email3 = 'abc@naver.com' # naver
email4 = 'abc@weniv.co.kr' # weniv
email5 = 'li.cat@weniv.co.kr' # weniv

def email(email):
    for a in range(1,len(email)):
        if email.index('@')<email.index('.',a):
            return print(email[email.index('@')+1:email.index('.',a)])
        else:
            continue
email(email2)
email(email3)
email(email4)
email(email5)
```
