```python
# 문제 1. 다음 함수 is_palindrome를 완성하시오.
# 앞으로 읽으나 뒤로 읽으나 같은 문자열이면 True, 아니면 False 반환
def is_palindrome(s):
    if s == s[::-1]:
        return True
    return False

print(is_palindrome("level"))  # 결과: True
print(is_palindrome("python")) # 결과: False
```

```python
# 문제 2.리스트 numbers가 주어졌을 때, 이 리스트 내의 모든 짝수를 제거하는 코드를 작성하시오.
numbers = [2, 3, 4, 5, 6, 7, 8, 9, 10]
numbers = list(filter(lambda x:x%2==1, numbers))
print(numbers)  # 결과: [3, 5, 7, 9]
```
