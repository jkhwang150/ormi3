```python
# global을 사용하지 않고 중첩된 함수 내에서 a를 출력하게 되면 바로 위 함수의 a를
a = 10
def f():
    a = 100
    print(a)
    def ff():
        a = 1000
        print(a)
        def fff():
            print(a)
        fff()
    ff()
f()
print(a)
```

```python
# nonlocal
a = 10
def f():
    a = 100
    print(f'f a: {a}')
    def ff():
        a = 1000
        print(f'ff a: {a}')
        def fff():
            nonlocal a
            a = 100
            print(f'fff a: {a}')
        fff()
        print(f'ff a: {a}')
    ff()
f()
print(f'global a: {a}')
```

```python
def factorial(n):
    if n <= 1:
        return 1
    return n*factorial(n-1)

factorial(6)
```

```python
def reversed_string(n):
    if len(n) <= 1:
        return n
    return reversed_string(n[1:]) + n[0]
reversed_string('hello world')
```

```python
data = [
    [2, 'hello world', 30],
    [3, 'hworl', 33],
    [4, 'hello w', 31],
    [5, 'hello wo', 35],
    [6, 'hello w', 39]
]
```

```python
# 1. 첫 번째 요소의 값으로 정렬하시오.
sorted(data)

# 2. 두 번째 요소의 문자열 길이를 기준으로 정렬하시오.
sorted(data, key = lambda x:len(x[1]))

# 3. 세 번째 요소의 문자열 길이를 기준으로 정렬하시오.
sorted(data, key = lambda x:x[2])

# 4. 첫 번째 요소와 세 번째 요소의 차가 큰 순으로 출력하시오.
sorted(data, key = lambda x:x[2]-x[0], reverse = True)
```
