```python
#1. 주어진 리스트 data = [{"name": "길동", "수학": 3, "과학": 93}, {"name": "춘향", "수학": 33, "수학": 11}, {"name": "철수", "수학": 94, "과학": 67}]에서 평균점수가 가장 높은 학생에 이름을 출력하세요.

#2. 주어진 사전 grades = {"Tom": 87, "Jerry": 95, "Mickey": 70}의 모든 값을 5점씩 증가시키고 결과를 출력하세요.

#3. 주어진 문자열 s = "apple banana apple cherry banana cherry apple"에서 가장 자주 등장하는 단어를 출력하세요.
```

```python
# 문제 1.주어진 리스트에서 평균점수가 가장 높은 학생을 출력하시오
data = [{"name": "길동", "수학": 3, "과학": 93}, {"name": "춘향", "수학": 33, "과학": 11}, {"name": "철수", "수학": 94, "과학": 67}]
grade = []

for i in data:
    value = list(filter(lambda x:type(x) is int, i.values()))
    avg = sum(value) / len(value)
    grade.append(avg)

data[grade.index(max(grade))]
```

```python
# 문제 2. 주어진 사전의 모든 값을 5점씩 증가시키고 결과를 출력하세요.
grades = {"Tom": 87, "Jerry": 95, "Mickey": 70}
key = [i for i in grades.keys()]
for i in key:
    grades[i] += 5
print(grades)
```

```python
# 문제 3. 주어진 문자열에서 가장 자주 등장하는 단어를 출력하세요.
s = "apple banana apple cherry banana cherry apple"
lis = s.split(" ")
lis_count=[]
for i in lis:
    lis_count.append(lis.count(i))
freq = lis_count.index(max(lis_count))
answer = lis[freq]
```
