```python
국영수 = {'국어': 97, '영어': 38, '수학':100}
# 문제1: 다음 국영수 dict에 점수에 평균을 구하는 코드를 작성하세요.
# 문제2: {'사회':70, '과학':85}가 추가되었습니다. dict에 평균을 구하는 코드를 작성하세요.
# 문제3: 영어와 수학점수가 각각 10점씩 올랐습니다. 점수를 10점 추가해주세요.

# 문제 1. dict에 점수 평균
국영수.values()
a = sum(국영수.values())
b = len(국영수.values())
avg = a/b
print(avg)

# 문제 2. 사회,과학 성적을 추가하여 dict에 평균을 구하시오.
국영수.update({'사회':70, '과학':85})
c = sum(국영수.values())
d = len(국영수.values())
avg2 = c/d
print(avg2)

# 문제 3. 영어와 수학점수가 10점씩 오른것을 추가하시오.
국영수['영어'] = 48
국영수['수학'] = 110
print(국영수)

# 문제 4. 최종점수에서 가장 높은 점수를 출력하시오.
high_grade = max(국영수.values())
low_grade = min(국영수.values())
print(high_grade, low_grade)
```
