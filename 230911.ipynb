# 매서드 채이닝

```python
# 연속으로 메설드를 호출하는 것을 말한다.
# 문자열 메서드 채이닝
# 코드의 가독성을 해칠 수 있으니 많이 사용 X
sentence = "  Hello, World!  "
result = sentence.strip().lower().replace("world", "Python") # 공백제거, 소문자변환, 문자열변환 순서로 진행
print(result)  # "hello, Python!"
```

```python
# bool -> int
true_int = int(True)
false_int = int(False)

print(true_int)  # 출력: 1
print(false_int)  # 출력: 0
```

```python
# bool 형변환
print("bool('test'):", bool('test'))
print("bool(' '):", bool(' '))
print("bool(''):", bool(''))
print("bool(100):", bool(100))
print("bool(1):", bool(1))
print("bool(0):", bool(0))
print("bool(-1):", bool(-1))
print("bool(None):", bool(None))
print("bool([]):", bool([])) # 주의 (js에서는 true, Python에서 False)
print("bool(tuple()):", bool(tuple())) # 주의
print("bool({}):", bool({})) #주의
print("bool(set()):", bool(set())) #주의
```

```python
# 실무에서 많이 사용
x = [1,2,3]
while x:
    print(x)
    x.pop()

x = [1,2,3]
while x:
    print(x.pop())
```

```python
# set형으로 형변환 - 중복제거시 많이 사용
s = 'weniv CEO licat'
print(set(s)) #중복을 허락하지 않고 순서가 없습니다.
print(len(set(s))) #중복을 제거한 상태로의 길이를 알려줍니다.
print(len(s))
```
